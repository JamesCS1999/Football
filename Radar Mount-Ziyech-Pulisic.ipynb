```python
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
from soccerplots.utils import add_image
```

```python
mount_values = [0.27, 0.10, 1.88, 0.19, 0.28, 6.88]
pulisic_values = [0.29, 0.07, 1.39, 0.05, 0.13, 7.4]
ziyech_values = [0.24, 0.11, 2.15, 0.35, 0.32, 6.22]
```

```python
# parameter names of the statistics we want to show
params = ["npxG",  "xA/Key pass",'PPA', "Through Balls","Shot-Creating Dribbles", 'Progressive carries']

# The lower and upper boundaries for the statistics
low =  [0.18, 0.05, 0.54, 0.04, 0.07, 3.67]
high = [0.59, 0.18, 2.7, 0.44, 0.4, 9.25]
```

```python
# create lists of lists for player values

values = []
values.append(mount_values)
values.append(pulisic_values)
values.append(ziyech_values)
```

```python
# create list of tuples from low/high range lists

ranges = list(zip(low,high))
```

## Mount (Red), Pulisic (Blue), Ziyech (Green)

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges=ranges,params=params,values=values,
                         radar_color=['red','blue','green'],
                         alphas=[.75,.6,.5], image='Chelsea.png', image_coord=[0.465, 0.9, 0.1, 0.075],
                         compare=True)
```
